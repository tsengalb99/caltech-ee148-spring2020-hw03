```python
import time
import numpy as np

from MulticoreTSNE import MulticoreTSNE as TSNE

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

import multiprocessing as mp
from multiprocessing import Pool
from PIL import Image
import os
```

```python
data = np.load('../models/mnist/cus_final_vis/feats.npy', allow_pickle=True)
```

```python
feats = []
pred = []
real = []
pics = []
for i in range(len(data)):
    feats.append(data[i][2])
    pred.append(int(data[i][0]))
    real.append(int(data[i][1]))
    pics.append(data[i][3])
feats = np.array(feats)
print(feats.shape)
```

```python
X_embedded = TSNE(n_jobs=32, n_components=2).fit_transform(feats)
```

```python
colors = ['red', 'green', 'blue', 'orange', 'yellow', 'black', 'lavender', 'purple', 'brown', 'pink']
```

```python
plt_cols = [colors[real[i]] for i in range(len(X_embedded))]
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], color=plt_cols)
```

```python
p = Pool(32)
close = [1, 10, 100, 1000]
for choice in close:
    os.makedirs(str(choice), exist_ok=True)
for choice in close:
    data = [feats[_] - feats[choice] for _ in range(len(feats))]
    dists = p.map(np.linalg.norm, data)
    dists = [(dists[i], i) for i in range(len(dists))]
    dists = sorted(dists)
    for i in range(9):
        which = dists[i][1]
        Image.fromarray(pics[which]).save(os.path.join(str(choice), '{}.png'.format(i)))
```
